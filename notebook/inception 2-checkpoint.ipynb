```python
# ================= STEP 1: BASIC SETUP =================

import os
import cv2
import json
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

print("TensorFlow version:", tf.__version__)
print("GPU devices:", tf.config.list_physical_devices('GPU'))

# Dataset paths (Windows)
DATASET_PATH = r"E:\new segmented"
METADATA_PATH = r"E:\patch_metadata.json"

print("\nDataset path exists:", os.path.exists(DATASET_PATH))
if os.path.exists(DATASET_PATH):
    print("Folders inside dataset:", os.listdir(DATASET_PATH))
```

```python
def leaf_pixel_percentage(patch):
    """
    Estimates how much of the patch is leaf tissue (green).
    """
    hsv = cv2.cvtColor(patch, cv2.COLOR_RGB2HSV)

    lower_green = np.array([25, 40, 40])
    upper_green = np.array([90, 255, 255])

    mask = cv2.inRange(hsv, lower_green, upper_green)
    green_pixels = np.sum(mask > 0)
    total_pixels = patch.shape[0] * patch.shape[1]

    return (green_pixels / total_pixels) * 100
```

```python
def filter_patches_leaf_dominant(patches,
                                 max_black=15,
                                 min_green=40):
    """
    FINAL journal-grade filter:
    - ≤15% near-black pixels
    - ≥40% green (leaf) pixels
    """
    clean_patches = []
    for p in patches:
        black_pct = calculate_black_pixel_percentage(p)
        green_pct = leaf_pixel_percentage(p)

        if black_pct <= max_black and green_pct >= min_green:
            clean_patches.append(p)

    return clean_patches
```

```python
# ================= REQUIRED FUNCTIONS (MUST RUN) =================

import cv2
import numpy as np

def split_image_into_patches(image, patch_size=32):
    h, w = image.shape[:2]
    h = (h // patch_size) * patch_size
    w = (w // patch_size) * patch_size
    image = cv2.resize(image, (w, h))

    patches = []
    for i in range(0, h, patch_size):
        for j in range(0, w, patch_size):
            patches.append(image[i:i+patch_size, j:j+patch_size])
    return patches


def calculate_black_pixel_percentage(patch, threshold=20):
    gray = cv2.cvtColor(patch, cv2.COLOR_RGB2GRAY)
    black_pixels = np.sum(gray < threshold)
    total_pixels = patch.shape[0] * patch.shape[1]
    return (black_pixels / total_pixels) * 100


def leaf_pixel_percentage(patch):
    hsv = cv2.cvtColor(patch, cv2.COLOR_RGB2HSV)

    lower_green = np.array([25, 40, 40])
    upper_green = np.array([90, 255, 255])

    mask = cv2.inRange(hsv, lower_green, upper_green)
    green_pixels = np.sum(mask > 0)
    total_pixels = patch.shape[0] * patch.shape[1]

    return (green_pixels / total_pixels) * 100


def filter_patches_leaf_dominant(patches, max_black=15, min_green=40):
    clean = []
    for p in patches:
        if (calculate_black_pixel_percentage(p) <= max_black and
            leaf_pixel_percentage(p) >= min_green):
            clean.append(p)
    return clean
```

```python
# NEXT STEP: Define create_patch_metadata_full()

import os
import json
from tqdm import tqdm

def create_patch_metadata_full(data_dir, save_path):
    metadata = []
    stats = {
        "healthy": 0,
        "unhealthy": 0,
        "total_images": 0,
        "total_patches": 0
    }

    print("Starting FULL dataset processing...")
    print("Dataset path:", data_dir)

    folders = [
        f for f in os.listdir(data_dir)
        if os.path.isdir(os.path.join(data_dir, f))
    ]

    for folder in tqdm(folders, desc="Processing folders"):
        folder_path = os.path.join(data_dir, folder)

        # Label rule
        label = 0 if "healthy" in folder.lower() else 1

        for img_name in os.listdir(folder_path):
            img_path = os.path.join(folder_path, img_name)

            img = cv2.imread(img_path)
            if img is None:
                continue

            stats["total_images"] += 1

            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            patches = split_image_into_patches(img)
            patches = filter_patches_leaf_dominant(patches)

            for idx in range(len(patches)):
                metadata.append({
                    "image_path": img_path,
                    "patch_index": idx,
                    "label": label
                })

                stats["total_patches"] += 1
                if label == 0:
                    stats["healthy"] += 1
                else:
                    stats["unhealthy"] += 1

    with open(save_path, "w") as f:
        json.dump({"metadata": metadata, "stats": stats}, f, indent=2)

    print("Metadata saved to:", save_path)
    return metadata, stats
```

```python
# ================= STEP 4: TRAIN / VALIDATION SPLIT =================

from sklearn.model_selection import train_test_split

with open(METADATA_PATH, "r") as f:
    data = json.load(f)

metadata = data["metadata"]

train_meta, val_meta = train_test_split(
    metadata,
    test_size=0.2,
    random_state=42,
    shuffle=True
)

print("Training patches   :", len(train_meta))
print("Validation patches :", len(val_meta))

```

```python
# ================= STEP 6: PATCH DATA GENERATOR =================

class PatchDataGenerator(tf.keras.utils.Sequence):

    def __init__(self, metadata, batch_size=32, patch_size=32, shuffle=True):
        self.metadata = metadata
        self.batch_size = batch_size
        self.patch_size = patch_size
        self.shuffle = shuffle
        self.indexes = np.arange(len(metadata))
        self.on_epoch_end()

    def __len__(self):
        return int(np.ceil(len(self.metadata) / self.batch_size))

    def on_epoch_end(self):
        if self.shuffle:
            np.random.shuffle(self.indexes)

    def __getitem__(self, idx):
        batch_ids = self.indexes[idx*self.batch_size:(idx+1)*self.batch_size]

        X, y = [], []
        cache = {}

        for i in batch_ids:
            meta = self.metadata[i]
            path = meta["image_path"]
            patch_idx = meta["patch_index"]
            label = meta["label"]

            if path not in cache:
                img = cv2.imread(path)
                if img is None:
                    continue

                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                patches = split_image_into_patches(img)
                patches = filter_patches_leaf_dominant(patches)
                cache[path] = patches

            if patch_idx < len(cache[path]):
                X.append(cache[path][patch_idx])
                y.append(label)

        X = np.array(X, dtype=np.float32) / 255.0
        y = np.array(y, dtype=np.int32)

        return X, y
```

```python
train_gen = PatchDataGenerator(train_meta, batch_size=32, shuffle=True)
val_gen   = PatchDataGenerator(val_meta, batch_size=32, shuffle=False)

print("Train batches:", len(train_gen))
print("Val batches  :", len(val_gen))
```

```python
# ================= STEP 7A: SMALL INCEPTION MODEL =================

from tensorflow.keras import layers, models

def conv_module(x, filters, kernel_size, strides=1, padding="same"):
    x = layers.Conv2D(filters, kernel_size,
                      strides=strides,
                      padding=padding,
                      use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)
    return x


def inception_module(x,
                     f1,
                     f3r, f3,
                     f5r, f5,
                     fpool):
    # 1x1
    b1 = conv_module(x, f1, 1)

    # 3x3
    b3 = conv_module(x, f3r, 1)
    b3 = conv_module(b3, f3, 3)

    # 5x5
    b5 = conv_module(x, f5r, 1)
    b5 = conv_module(b5, f5, 5)

    # Pool
    bp = layers.MaxPooling2D(3, strides=1, padding="same")(x)
    bp = conv_module(bp, fpool, 1)

    return layers.Concatenate()([b1, b3, b5, bp])


def build_small_inception(input_shape=(32, 32, 3)):
    inputs = layers.Input(shape=input_shape)

    x = conv_module(inputs, 64, 3)

    x = inception_module(x,
                          f1=32,
                          f3r=32, f3=64,
                          f5r=16, f5=32,
                          fpool=32)

    x = layers.Conv2D(128, 3, strides=2, padding="same")(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)

    x = inception_module(x,
                          f1=64,
                          f3r=64, f3=128,
                          f5r=32, f5=64,
                          fpool=64)

    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dropout(0.4)(x)
    x = layers.Dense(128, activation="relu")(x)
    x = layers.Dropout(0.3)(x)

    outputs = layers.Dense(1, activation="sigmoid")(x)

    model = models.Model(inputs, outputs, name="Small_Inception_PatchNet")
    return model


model = build_small_inception()
model.summary()
```

```python
# ================= STEP 7B: COMPILE + TRAIN =================

from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

model.compile(
    optimizer=Adam(learning_rate=1e-3),
    loss="binary_crossentropy",
    metrics=[
        "accuracy",
        tf.keras.metrics.Precision(name="precision"),
        tf.keras.metrics.Recall(name="recall")
    ]
)

callbacks = [
    EarlyStopping(
        monitor="val_loss",
        patience=5,
        restore_best_weights=True
    ),
    ReduceLROnPlateau(
        monitor="val_loss",
        factor=0.5,
        patience=3,
        min_lr=1e-6
    )
]

history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=25,
    
    callbacks=callbacks,
    verbose=1
)
```

```python
# # STEP 0: Define Simple CNN model

from tensorflow.keras import layers, models

def build_simple_cnn(input_shape=(32, 32, 3)):
    model = models.Sequential([
        layers.Input(shape=input_shape),

        layers.Conv2D(32, (3,3), activation="relu", padding="same"),
        layers.MaxPooling2D(2,2),

        layers.Conv2D(64, (3,3), activation="relu", padding="same"),
        layers.MaxPooling2D(2,2),

        layers.Conv2D(128, (3,3), activation="relu", padding="same"),
        layers.GlobalAveragePooling2D(),

        layers.Dense(128, activation="relu"),
        layers.Dropout(0.4),
        layers.Dense(1, activation="sigmoid")
    ])

    return model

```

```python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras import layers, models

def build_mobilenet_v2(input_shape=(32, 32, 3)):
    base_model = MobileNetV2(
        input_shape=input_shape,
        include_top=False,
        weights=None    # IMPORTANT: fair comparison
    )

    model = models.Sequential([
        base_model,
        layers.GlobalAveragePooling2D(),
        layers.Dense(128, activation="relu"),
        layers.Dropout(0.3),
        layers.Dense(1, activation="sigmoid")
    ])

    return model
```
